```python
import numpy as np
import matplotlib.pyplot as plt

# Normal Distribution
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Uniform Distribution
uniform_data = np.random.uniform(low=0, high=1, size=1000)

# Plotting
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(normal_data, bins=30)
plt.title("Normal Distribution")

plt.subplot(1,2,2)
plt.hist(uniform_data, bins=30)
plt.title("Uniform Distribution")

plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

# Load dataset
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build model
model = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='sgd',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(x_train, y_train, epochs=5)

# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test)

print("Test Accuracy:", test_acc)
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

# Load dataset
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

x_train = x_train / 255.0
x_test = x_test / 255.0

# Model
model = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
model.fit(x_train, y_train, epochs=3)

# Evaluate
loss, accuracy = model.evaluate(x_test, y_test)

print("Validation Loss:", loss)
print("Validation Accuracy:", accuracy)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# True parameters of Normal Distribution
true_mean = 5
true_std = 2

# Different dataset sizes
sample_sizes = [10, 50, 100, 500, 1000]

results = []

for size in sample_sizes:

    # Generate synthetic dataset
    data = np.random.normal(loc=true_mean, scale=true_std, size=size)

    # MLE Estimates
    estimated_mean = np.mean(data)
    estimated_std = np.std(data)

    # Estimation Errors
    mean_error = abs(true_mean - estimated_mean)
    std_error = abs(true_std - estimated_std)

    # Store results
    results.append([
        size,
        estimated_mean,
        estimated_std,
        mean_error,
        std_error
    ])

# Create DataFrame
df = pd.DataFrame(results, columns=[
    "Sample Size",
    "Estimated Mean",
    "Estimated Std Dev",
    "Mean Error",
    "Std Dev Error"
])

print(df)

# Plot Errors
plt.figure(figsize=(8,5))

plt.plot(sample_sizes, df["Mean Error"], marker='o', label="Mean Error")
plt.plot(sample_sizes, df["Std Dev Error"], marker='o', label="Std Dev Error")

plt.xlabel("Sample Size")
plt.ylabel("Error")
plt.title("MLE Estimation Error vs Sample Size")
plt.legend()

plt.show()
```
